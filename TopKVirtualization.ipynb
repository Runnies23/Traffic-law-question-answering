```python
pattern  = r"([^-\n]+) - "
```

```python
def topk1(dataframe):#topk 1
  performance  = 0
  topk1 = dataframe['col_0']
  topk1_df = []
  for i in topk1:
      i = str(i)
      matches = re.findall(pattern,i)
      topk1_df.append(matches)
  newdfBM25_df = pd.DataFrame({"top1":topk1_df})
  newdftop1 = newdfBM25_df['top1'].str[0].str.replace(r"\[\['|'\]", '', regex=True).replace(r"^\['", '', regex=True).replace('^\["', "", regex=True)
  newdftopk = pd.DataFrame({"top1": newdftop1})
  for index in range(len(newdftop1)):
      if newdftopk["top1"][index] == answer_df["REALANS"].iloc[index]:
          performance += 1
  print(performance)
  print(f"percentage : {performance/103} %")
  return performance
```
